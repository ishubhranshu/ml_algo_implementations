```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/ipl.csv")
df.sample(5)
```

```python
df.columns
```

```python
cols = ['season', 'city', 'team1', 'team2', 'toss_winner', 'toss_decision', 'winner']

df = df[cols]
df.sample(5)
```

```python
df.info()
```

```python
df.dropna(inplace=True)
# df['season'] = df['season'].astype('object')
df.info()
```

```python
encoders = {}

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for col in df.columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le
```

```python
encoders
```

```python
df.sample(5)
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state = 42)


print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
from sklearn.naive_bayes import CategoricalNB, GaussianNB, BernoulliNB, MultinomialNB

cnb = CategoricalNB()

cnb.fit(xtrain, ytrain)
```

```python
ypred = cnb.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(ytest, ypred)
```
