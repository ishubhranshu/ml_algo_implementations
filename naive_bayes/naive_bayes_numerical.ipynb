```python
# import required libraries
import pandas as pd
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB, CategoricalNB, MultinomialNB, ComplementNB
```

```python
# load data
df = pd.read_csv('../../data/Naive-Bayes-Classification-Data.csv')
print(df.shape)
df.head(5)
```

```python
# split input and output column
x = df.iloc[:, :2]
y = df.iloc[:, 2]
print(x.shape)
print(y.shape)
```

```python
# split training and testing data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42)

print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
# plot distributions
sns.pairplot(df[df['diabetes'] == 0].iloc[:, :2])
```

```python
# plot distribution
sns.pairplot(df[df['diabetes'] == 1].iloc[:, :2])
```

```python
# load models
gnb = GaussianNB()
bnb = BernoulliNB()
cnb = CategoricalNB()
mnb = MultinomialNB()
ccnb = ComplementNB()

# train models
gnb.fit(xtrain, ytrain)
bnb.fit(xtrain, ytrain)
cnb.fit(xtrain, ytrain)
mnb.fit(xtrain, ytrain)
ccnb.fit(xtrain, ytrain)
```

```python
# prediction
gnbpred = gnb.predict(xtest)
bnbpred = bnb.predict(xtest)
cnbpred = cnb.predict(xtest)
mnbpred = mnb.predict(xtest)
ccnbpred = ccnb.predict(xtest)
```

```python
# evaluation
print('G', accuracy_score(ytest, gnbpred))
print('B', accuracy_score(ytest, bnbpred))
print('C', accuracy_score(ytest, cnbpred))
print('M', accuracy_score(ytest, mnbpred))
print('CC', accuracy_score(ytest, ccnbpred))
```
