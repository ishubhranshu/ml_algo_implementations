```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
traindf = pd.read_csv("mnist_train.csv")
print(traindf.shape)
traindf. sample(5)
```

```python
testdf = pd.read_csv("mnist_test.csv")
print(testdf.shape)
testdf. sample(5)
```

```python
xtrain, xtest, ytrain, ytest  = traindf.iloc[:, 1:], testdf.iloc[:, 1:], traindf.iloc[:, 0], testdf.iloc[:, 0]

print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=15)

clf.fit(xtrain, ytrain)
```

```python
ypred = clf.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(ytest, ypred)
acc
```

```python
from sklearn.model_selection import GridSearchCV

# define  parameter and possible values
param_dist = {
    'criterion':['gini', 'entropy'],
    'max_depth':[1,2,3,4,5,None],
    'splitter':['best', 'random'],
    'random_state':[0,42, None]
}

# fit gridCV
grid = GridSearchCV(clf, param_grid=param_dist, cv=10, n_jobs=-1)
```

```python
#train gridCV
grid.fit(xtrain, ytrain)
```

```python
# best score
grid.best_score_
```
