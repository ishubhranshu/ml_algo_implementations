```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Generate x values
x = np.linspace(0, 1000, 200)  # 100 points between 0 and 10

# Define slope and intercept
slope = 2
intercept = 1

# Generate y values with some noise
y = slope * x + intercept + np.random.normal(0, 400, 200)  # Add random noise

# Create a pandas DataFrame if desired
df = pd.DataFrame({'x': x, 'y': y})

print(df)
sns.scatterplot(x=x, y=y)
```

```python
x = df[['x']]
y = df['y']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42)

print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(xtrain, ytrain)
```

```python
ypred = lr.predict(xtest)
ypred
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(ytest, ypred)
```

```python
print(lr.coef_[0])
print(lr.intercept_)
```

```python
# Plotting the data points and regression line
plt.scatter(xtrain, ytrain, color='blue', label='Training Data', alpha=0.3, s=20)  # Training data points
plt.scatter(xtest, ytest, color='red', label='Test Data', alpha=0.3, s=20)  # Test data points
plt.plot(xtrain, lr.predict(xtrain), color='black', label='Regression Line')  # Regression line
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
